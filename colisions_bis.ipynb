```python
from IPython.display import display, HTML
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from ipywidgets import interact, FloatSlider

class Boule:
    def __init__(self, rayon, Xcentre, Ycentre, angle, vitesse, couleur):
        # Constructeur pour initialiser les attributs
        self.angle = angle
        self.vitesse = vitesse
        self.rayon = rayon
        self.Xcentre = Xcentre
        self.Ycentre = Ycentre
        self.couleur = couleur
        self.Xvitesse = vitesse * (np.cos(angle*np.pi/180))
        self.Yvitesse = vitesse * (np.sin(angle*np.pi/180))
        self.circle = plt.Circle((Xcentre, Ycentre), rayon, fc=couleur)

    def afficher(self):
        ax.add_patch(self.circle)
    
    def avancer(self):
        if self.Xcentre-self.rayon<0 or self.Xcentre+self.rayon>width:
            self.Xvitesse=-self.Xvitesse
        if self.Ycentre-self.rayon<0 or self.Ycentre+self.rayon>height:
            self.Yvitesse=-self.Yvitesse
        self.Xcentre=self.Xcentre+self.Xvitesse
        self.Ycentre=self.Ycentre+self.Yvitesse
        (self.circle).set_center((self.Xcentre, self.Ycentre))
        
# Paramètres de la fenêtre 
width, height = 20, 20  # Dimensions de la fenêtre

# Création de la figure et des axes
fig, ax = plt.subplots()
#ax.set_title("")
ax.set_xlim(0, width)
ax.set_ylim(0, height)
ax.set_aspect('equal')
ax.set_xticks([]) 
ax.set_yticks([]) 
fig.texts.clear()

# Création d'un objet de type Boule
boule1 = Boule(1,1,1,65,0.8,'blue')
boule2 = Boule(2,10,8,-35,0.4,'red')
# Fonction de mise à jour pour l'animation
def update(frame):
    #global boule1
    boule1.afficher()
    boule2.afficher()
    boule1.avancer()
    boule2.avancer()
    return 0,

# Création de l'animation
ani = FuncAnimation(fig, update, frames=1000, interval=20, blit=True)


# Widget interactif
#interact(set_speed, new_speed=FloatSlider(value=2, min=0.1, max=5, step=0.1))
plt.show()
```
